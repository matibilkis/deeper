```python
import os 
import tensorflow as tf
import numpy as np
from nets import RNNC
from tqdm import tqdm
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2

name="task2_dset"
data, labels = np.load(name+"/data.npy"), np.load(name+"/label.npy")

net = RNNC()
print("data[:4]: \n",data[:4], "\n")
print("labels[:4]\n{}\n".format(labels[:4]))
l1 =tf.losses.mean_squared_error(labels,tf.squeeze(net(data)))
print("l1[:4]: \n{}\n".format(l1[:4]))
l2 = tf.losses.mean_squared_error(np.expand_dims(labels,axis=2),net(data))
print("l2[:4]: \n{}\n".format(l2[:4]))
print("mean_l1: {}".format(tf.reduce_mean(l1)))
print("mean_l2: {}".format(tf.reduce_mean(l2)))
```

```python
stoppings = []
for k in range(4):
    for kk in range(10**k,10**(k+1),10**k):
        stoppings.append(kk)
for j in range(1,4):
    stoppings.append(j*10**4)
```

```python
for lmm in [0,1]:

    net = RNNC(seed_val=0.5,lmode=lmm)

    l=[]
    for k in tqdm(range((3*10**2)+1)):
        if k in stoppings:
            lo, gr = net.train_step(data, labels, clipping=True,return_gradients=True)
            np.save("grads_{}/{}".format(lmm,k),gr)
            net.save_weights("net_{}/".format(lmm)+str(k))        
            l.append(lo)
        else:
            l.append(net.train_step(data, labels, clipping=True))
    os.system("python3 image_gen.py --names {}".format(lmm))
    means=[]
    maxs=[]
    for ep in stoppings:
        gr = np.load("grads_{}/{}.npy".format(lmm,ep), allow_pickle=True)
        means.append([np.mean(gr[k].numpy()) for k in range(len(gr))])
        maxs.append([np.max(np.abs(gr[k].numpy())) for k in range(len(gr))])    


    ti=[]
    for k in np.log10(stoppings):
        if k in list(range(10)):
            ti.append(k)

    plt.figure(figsize=(10,30))
    axs={}
    for k in range(9):
        axs[str(k)] = plt.subplot2grid((9,2),(k,0))
        axs[str(k)].plot(np.log10(stoppings),np.array(means)[:,k], label="mean gradient at layer {}".format(k), color="red",alpha=0.75, linewidth=4)
        axs[str(k)].set_xticks(ti)    
        axs[str(k)].set_xticklabels(["10^{}".format(int(kk)) for kk in ti])
        if k==0:
            axs[str(k)].set_title("mean value of gradient \nat each layer", size=20)
        axs[str(k)] = plt.subplot2grid((9,2),(k,1))
        axs[str(k)].plot(np.log10(stoppings),np.array(maxs)[:,k], label="max absolute value of gradient at layer {}".format(k), color="red",alpha=0.75, linewidth=4)
        axs[str(k)].set_xticks(ti)    
        axs[str(k)].yaxis.tick_right()
        axs[str(k)].set_xticklabels(["10^{}".format(int(kk)) for kk in ti])
        if k==0:
            axs[str(k)].set_title("max abs value of gradient \nat each layer", size=20)
    plt.savefig("evolution_lmm_{}.png".format(lmm))
```
