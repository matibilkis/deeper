```python
import os 
import tensorflow as tf
import numpy as np
from nets import RNNC
from tqdm import tqdm
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2

data, labels = np.load("ProfilesSet/data.npy"), np.load("ProfilesSet/label.npy")
dat_set = tf.data.Dataset.from_tensor_slices((data,labels))

```

```python
b1 = -.5
betas = np.arange(-1,1,.05)
datfeed={0:[],1:[]}
for n1 in [0,1]:
    for b2 in betas:
        datfeed[n1].append([[b1,0],[b2,n1]])
for n in [0,1]:
    datfeed[n] = np.array(datfeed[n])
```

```python
plt.figure(figsize=(10,10))
plt.title("First feed of the whole \ndata to the (untrained) RNN net")
ax={}
for n in [0,1]:
    ax[n]=plt.subplot2grid((2,2),(0,n))
ax2 = plt.subplot2grid((2,2),(1,0),colspan=2)
for k in [0.01, 0.1, 1,10]:
    net = RNNC(return_sequences=False, seed_val=k)
    for n in [0,1]:
        ax[n].plot(np.squeeze(net(datfeed[n])), label="initial seed: {}".format(k), linewidth=4)
    ax2.plot(np.log10(np.abs(np.squeeze(net(datfeed[0])-net(datfeed[1])))), label="initial seed: {}".format(k), linewidth=4)
    ax2.set_ylabel("log10 difference of\n predictions for different n",size=20)
for n in [0,1]:
    ax[n].set_ylabel("prediction n={}".format(n),size=20)
    ax[n].legend()#plt.savefig("variations.png")
ax[n].yaxis.tick_right()
ax2.legend()
plt.savefig("different_initializations.png")
```
